```python
import torch

from ratsnlp.nlpbook.classification import ClassificationTrainArguments
args = ClassificationTrainArguments(
    pretrained_model_name='beomi/kcbert-base', #pretrain 마친 모델 이름
    downstream_corpus_name="nsmc", #downstream task에 사용될 데이터 이름
    downstream_model_dir='/gdrive/My Drive/nlpbook/checkpoint-doccls', #파인튜닝된 모델의 체크포인트가 저장될 위치
    batch_size=32 if torch.cuda.is_available() else 4,
    learning_rate=5e-5,
    max_seq_length=128,
    epochs=3,
    tpu_cores=0 if torch.cuda.is_available() else 8,
    seed=7
)
```

```python
from ratsnlp import nlpbook

nlpbook.set_seed(args) #시드 고정
```

```python
nlpbook.set_logger(args)
```

```python
from Korpora import Korpora


Korpora.fetch(
    corpus_name=args.downstream_corpus_name,
    root_dir=args.downstream_corpus_root_dir,
    force_download=True,
)
```

```python
args.downstream_corpus_root_dir
```

```python
from transformers import BertTokenizer

tokenizer=BertTokenizer.from_pretrained(
    args.pretrained_model_name,
    do_lower_case = False,
)
```

```python
#학습 데이터 구축
from ratsnlp.nlpbook.classification import NsmcCorpus, ClassificationDataset 

corpus = NsmcCorpus()
train_dataset= ClassificationDataset(
    args=args,
    corpus=corpus,
    tokenizer=tokenizer,
    mode="train",
)
```

```python
train_dataset[0]
```

```python
#학습 Dataloader 구출
from torch.utils.data import DataLoader, RandomSampler

train_dataloader = DataLoader(
    train_dataset,
    batch_size=args.batch_size,
    sampler=RandomSampler(train_dataset, replacement=False),
    collate_fn=nlpbook.data_collator, #뽑아낸 인스턴스를 배치로 만드는 것
    drop_last=False,
    num_workers=args.cpu_workers,
)
```

```python
# test dataloader 구축
#sequential sampler은 인스턴스를 batch size만큼 순서대로 추출
from torch.utils.data import SequentialSampler

val_dataset = ClassificationDataset(
    args=args,
    corpus=corpus,
    tokenizer = tokenizer,
    mode="test",
)

val_dataloader = DataLoader(
    val_dataset,
    batch_size=args.batch_size,
    sampler = SequentialSampler(val_dataset),
    collate_fn=nlpbook.data_collator,
    drop_last=False,
    num_workers=args.cpu_workers,
)
```

```python
#모델 불러오기
from transformers import BertConfig, BertForSequenceClassification

pretrained_model_config = BertConfig.from_pretrained(
    args.pretrained_model_name,
    num_labels=corpus.num_labels,
)

model = BertForSequenceClassification.from_pretrained(
    args.pretrained_model_name,
    config=pretrained_model_config,
)
```

```python
#모델 학습
#pytorch lightning의 lightning module class을 상속받아 task을 정의
from ratsnlp.nlpbook.classification import ClassificationTask

task = ClassificationTask(model, args)
```

```python
trainer = nlpbook.get_trainer(args)
```

```python
torch.cuda.device_count()
```

```python
torch.cuda.is_available()
```

```python
trainer.fit(
    task,
    train_dataloader=train_dataloader,
    val_dataloaders=val_dataloader,
)
```

## Inference

```python
from ratsnlp.nlpbook.classification import ClassificationDeployArguments

args = ClassificationDeployArguments(
    pretrained_model_name = "beomi/kcbert-base",
    downstream_model_dir="/gdrive/My Drive/nlpbook/checkpoint-doccls",
    max_seq_length=128,
)
```

```python
#load checkpoint
import torch
fine_tuned_model_ckpt = torch.load(
    args.downstream_model_checkpoint_fpath,
    map_location=torch.device("cpu"),
)
```

```python
#load configureation of BERT when used in finetuning
from transformers import BertConfig
pretrained_model_config = BertConfig.from_pretrained(
    args.pretrained_model_name,
    num_labels=fine_tuned_model_ckpt["state_dict"]["model.classifier.bias"].shape.numel(),
)
```

```python
#initialize BERT model
from transformers import BertForSequenceClassification
model = BertForSequenceClassification(pretrained_model_config)
```

```python
# put checkpoints in model
model.load_state_dict({k.replace("model.", ""): v for k, v in fine_tuned_model_ckpt['state_dict'].items()})
```

```python
#turn model into evaluation mode
model.eval()
```

```python
#model inference 
# 문장을 토큰화한 뒤, input_id, attention_masks, token_type_ids 만들기
def inference_fn(sentence):
  inputs=tokenizer(
      [sentence],
      max_length=args.max_seq_length,
      padding="max_length",
      truncation=True,
  )
  
  with torch.no_grad(): # graident연산을 끌 때 사용. Useful for inference. Reduce memory consumption 
    outputs=model(**{k: torch.tensor(v) for k, v in inputs.items()})
    prob=outputs.logits.softmax(dim=1)
    positive_prob=round(prob[0][1].item(), 4)
    negative_prob=round(prob[0][0].item(), 4)
    pred="긍정 (positive)" if torch.argmax(prob) == 1 else "부정 (negative)"
  
  return {
      'sentence' : sentence,
      'prediction' : pred,
      'positive_data' : f"긍정 {positive_prob}",
      'negative_data' : f"부정 {negative_prob}",
      'positive_width' : f"{positive_prob * 100}%", #추후 웹페이지에서 긍정및부정 막대 길이 조정
      'negative_width' : f"{negative_prob * 100}%",
  }
```

```python
from ratsnlp.nlpbook.classification import get_web_service_app

app = get_web_service_app(inference_fn)
app.run()
```
